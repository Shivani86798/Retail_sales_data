Importing important libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Importing the csv dataset file


```python
df=pd.read_csv("/content/Diwali Sales Data.csv", encoding='unicode-escape')
```

Deleting empty columns

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

As the amount column has 12 null values so we have to remove these null values

```python
df.shape
```

```python
#deleting the null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
#changing data type of Amount column from Float to integer
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

looking at the columns

```python
df.columns
```

Lets start visualising the data now

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen )
```

*From above graph it is clear that most of the buyers are women and even the purchasing power of women is more than men*

```python
df.columns
```

AGE

```python
ax=sns.countplot( data=df, x='Age Group', hue='Gender')
```

```python
#Total Amount by Age-Group
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

*From above graph it is evident that most of buyers are from 26-35 age group*

```python
df.columns
```

STATE

```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#to set the width of bars
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Sales as per marital status
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')
```

*From above graph it is clear that the most of the buyers are married(Women) and they have high purchasing power*

OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_state, x='Occupation', y='Amount')
```

*From above graph it is evident that most of the buyers are working in IT sector, Healthcare and aviation*

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Product_Category', data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
sns.set(rc={'figure.figsize':(20,5)})
```

*From the above graph we can see that the most of the sold products are from food, clothing & Apparel and Electronics*

```python
sales_state=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(data=sales_state, x='Product_ID', y='Orders')
sns.set(rc={'figure.figsize':(20,3)})
```

CONCLUSION

*Married women of age group 26-35 yrs from UP, Maharashtra and Karnatka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
